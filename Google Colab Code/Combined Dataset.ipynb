# **DATASET GABUNGAN - GOOGLE COLAB CODE**

# **Import Preparation**

```python
import numpy as np
import pandas as pd
```

# **Read the Data**

```python
# Variabel Dependen Y
data_ctra = pd.read_csv('/content/data_ctra_per_bulan.csv')
data_indf = pd.read_csv('/content/data_indf_per_bulan.csv')
data_asii = pd.read_csv('/content/data_asii_per_bulan.csv')
data_bsde = pd.read_csv('/content/data_bsde_per_bulan.csv')
data_icbp = pd.read_csv('/content/data_icbp_per_bulan.csv')
data_klbf = pd.read_csv('/content/data_klbf_per_bulan.csv')
data_itmg = pd.read_csv('/content/data_itmg_per_bulan.csv')
data_jpfa = pd.read_csv('/content/data_jpfa_per_bulan.csv')
data_tlkm = pd.read_csv('/content/data_tlkm_per_bulan.csv')
data_ultj = pd.read_csv('/content/data_ultj_per_bulan.csv')
data_aces = pd.read_csv('/content/data_aces_per_bulan.csv')
data_tspc = pd.read_csv('/content/data_tspc_per_bulan.csv')
data_smar = pd.read_csv('/content/data_smar_per_bulan.csv')
data_smsm = pd.read_csv('/content/data_smsm_per_bulan.csv')
data_jrpt = pd.read_csv('/content/data_jrpt_per_bulan.csv')
data_duti = pd.read_csv('/content/data_duti_per_bulan.csv')
data_epmt = pd.read_csv('/content/data_epmt_per_bulan.csv')
data_smcb = pd.read_csv('/content/data_smcb_per_bulan.csv')
data_pwon = pd.read_csv('/content/data_pwon_per_bulan.csv')
data_jsmr = pd.read_csv('/content/data_jsmr_per_bulan.csv')

# Variabel Independen X
data_usdidr = pd.read_csv('/content/data_usdidr_per_bulan.csv')
data_birate = pd.read_csv('/content/BI_Rate.csv')
data_inflasi = pd.read_csv('/content/Inflasi_convert.csv')
```

# **Combining Monthly Data**

```python
data_gabungan = pd.merge(data_usdidr, data_birate, on='Date', how='inner')
data_gabungan = pd.merge(data_gabungan, data_inflasi, on='Date', how='inner')
data_gabungan.rename(columns={'Close': 'Close USDIDR'}, inplace=True)


data_gabungan['Date'] = pd.to_datetime(data_gabungan['Date'], format='%b %Y')


data_list = [
    (data_ctra, 'Close CTRA'),
    (data_indf, 'Close INDF'),
    (data_asii, 'Close ASII'),
    (data_bsde, 'Close BSDE'),
    (data_icbp, 'Close ICBP'),
    (data_klbf, 'Close KLBF'),
    (data_itmg, 'Close ITMG'),
    (data_jpfa, 'Close JPFA'),
    (data_tlkm, 'Close TLKM'),
    (data_ultj, 'Close ULTJ'),
    (data_aces, 'Close ACES'),
    (data_tspc, 'Close TSPC'),
    (data_smar, 'Close SMAR'),
    (data_smsm, 'Close SMSM'),
    (data_jrpt, 'Close JRPT'),
    (data_duti, 'Close DUTI'),
    (data_epmt, 'Close EPMT'),
    (data_smcb, 'Close SMCB'),
    (data_pwon, 'Close PWON'),
    (data_jsmr, 'Close JSMR')
]


for data, close_name in data_list:
    data['Date'] = pd.to_datetime(data['Date'], format='%b %Y')
    data_gabungan = pd.merge(data_gabungan, data, on='Date', how='inner')
    data_gabungan.rename(columns={'Close': close_name}, inplace=True)


data_gabungan['Date'] = data_gabungan['Date'].dt.strftime('%b %Y')
```
