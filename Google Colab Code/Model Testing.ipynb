# Import Necessary Libraries

```python
import zipfile
import os
import tensorflow as tf
import numpy as np
import joblib
from tensorflow import keras

print(tf.__version__)
print(keras.__version__)
```

# Extracting Model Files from ZIP Archive

```python
zip_path = '/content/model.zip'
extract_dir = '/content/model'
os.makedirs(extract_dir, exist_ok=True)
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)
print(f"Model berhasil diekstrak di: {extract_dir}")
```

# Loading Saved Model with TFSMLayer for Inference

```python
saved_model_path = '/content/model'
model = keras.layers.TFSMLayer(filepath=saved_model_path, call_endpoint='serving_default')
```

# Loading Pre-trained Scalers for Input and Output Data

```python
skala_X = joblib.load('/content/skala_X.pkl')
skala_Y = joblib.load('/content/skala_Y.pkl')
```

# Predicting Stock Prices Based on New Data Input

```python
data_baru = np.array([[16000, 4.50, 3.90]])#Close USD, BI Rate, Inflasi
data_baru_standarisasi = skala_X.transform(data_baru)
prediksi_standarisasi = model(data_baru_standarisasi)

if isinstance(prediksi_standarisasi, dict):
    prediksi_standarisasi = prediksi_standarisasi['output_0']

prediksi_standarisasi = prediksi_standarisasi.numpy()

prediksi = skala_Y.inverse_transform(prediksi_standarisasi)

saham = ['Close CTRA', 'Close INDF', 'Close ASII', 'Close BSDE', 'Close ICBP',
         'Close KLBF', 'Close ITMG', 'Close JPFA', 'Close TLKM', 'Close ULTJ',
         'Close ACES', 'Close TSPC', 'Close SMAR', 'Close SMSM', 'Close JRPT',
         'Close DUTI', 'Close EPMT', 'Close SMCB', 'Close PWON', 'Close JSMR']

hasil_prediksi = dict(zip(saham, prediksi[0]))

print("Prediksi Harga Saham Berdasarkan Data Baru:")
for stock, price in hasil_prediksi.items():
    print(f"{stock}: {price:.2f}")
```

# Import Necessary Libraries

```python
import os
import numpy as np
import joblib
import json
import zipfile
```

# Extract tfjs_model.zip

```python
zip_path = '/content/tfjs_model.zip'
extract_dir = '/content/tfjs_model'
os.makedirs(extract_dir, exist_ok=True)
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)
print(f"Model TFJS berhasil diekstrak di: {extract_dir}")
```

# Save JavaScript code to test the model

```python
test_script = """
const tf = require('@tensorflow/tfjs-node');
const fs = require('fs');

// Load model
async function loadModel() {
    const modelPath = 'file:///content/tfjs_model/model.json';
    const model = await tf.loadGraphModel(modelPath);
    return model;
}

// Load and preprocess input data
async function predict() {
    const model = await loadModel();

    // Load standardization scalers from JSON
    const skala_X = JSON.parse(fs.readFileSync('/content/skala_X.json'));
    const skala_Y = JSON.parse(fs.readFileSync('/content/skala_Y.json'));

    // Input data in standardization format
    const data_baru = [[15000, 5.75, 3.2]];
    const data_baru_standarisasi = data_baru.map((d, i) => {
        return d.map((value, idx) => {
            return (value - skala_X.mean[idx]) / skala_X.std[idx];
        });
    });

    const inputTensor = tf.tensor2d(data_baru_standarisasi, [1, data_baru[0].length]);
    let prediksi_standarisasi = model.predict(inputTensor);
    prediksi_standarisasi = await prediksi_standarisasi.array();

    // Reverse scaling to get the original scale prediction
    const prediksi = prediksi_standarisasi[0].map((p, i) => {
        return (p * skala_Y.std[i]) + skala_Y.mean[i];
    });

    const saham = [
        'Close CTRA', 'Close INDF', 'Close ASII', 'Close BSDE', 'Close ICBP',
        'Close KLBF', 'Close ITMG', 'Close JPFA', 'Close TLKM', 'Close ULTJ',
        'Close ACES', 'Close TSPC', 'Close SMAR', 'Close SMSM', 'Close JRPT',
        'Close DUTI', 'Close EPMT', 'Close SMCB', 'Close PWON', 'Close JSMR'
    ];

    const hasilPrediksi = saham.reduce((acc, stock, index) => {
        acc[stock] = prediksi[index].toFixed(2);
        return acc;
    }, {});

    console.log("Prediksi Harga Saham Berdasarkan Data Baru:");
    console.log(hasilPrediksi);
}

predict().catch(err => console.error("Error during prediction:", err));
"""
```

# Save the JavaScript code to a file

```python
with open("/content/test_tfjs_model.js", "w") as file:
    file.write(test_script)
```

# Convert scalers to JSON format

```python
scaler_X = joblib.load('/content/skala_X.pkl')
scaler_Y = joblib.load('/content/skala_Y.pkl')
```

```python
scaler_X_json = {"mean": scaler_X.mean_.tolist(), "std": scaler_X.scale_.tolist()}
scaler_Y_json = {"mean": scaler_Y.mean_.tolist(), "std": scaler_Y.scale_.tolist()}
with open("/content/skala_X.json", "w") as f:
    json.dump(scaler_X_json, f)
with open("/content/skala_Y.json", "w") as f:
    json.dump(scaler_Y_json, f)
```
